```python
# Animate the radar data for nearly 1 hr interval

files_1 = sorted(glob.glob('D:/College/UG-3/Sem 6/BTP/Dataset/June 1st 0_4hr/*'))
files_2 = sorted(glob.glob('D:/College/UG-3/Sem 6/BTP/Dataset/June 1st 4_8hr/*'))
files_3 = sorted(glob.glob('D:/College/UG-3/Sem 6/BTP/Dataset/June 1st 8_12hr/*'))
files_4 = sorted(glob.glob('D:/College/UG-3/Sem 6/BTP/Dataset/June 1st 12_16hr/*'))
files_5 = sorted(glob.glob('D:/College/UG-3/Sem 6/BTP/Dataset/June 1st 16_20hr/*'))
files_6 = sorted(glob.glob('D:/College/UG-3/Sem 6/BTP/Dataset/June 1st 20_24hr/*'))

def animate_1(nframe):
    plt.clf()
    radar_1 = pyart.io.read_nexrad_archive(files_1[nframe])
    display_1 = pyart.graph.RadarDisplay(radar_1)
    display_1.plot_ppi('reflectivity', vmin = -32, vmax=80, cmap='pyart_NWSRef', title = files_1[nframe], colorbar_label='', axislabels_flag=False)

def animate_2(nframe):
    plt.clf()
    radar_2 = pyart.io.read_nexrad_archive(files_2[nframe])
    display_2 = pyart.graph.RadarDisplay(radar_2)
    display_2.plot_ppi('reflectivity', vmin = -32, vmax=80, cmap='pyart_NWSRef', title = files_2[nframe], colorbar_label='', axislabels_flag=False)

def animate_3(nframe):
    plt.clf()
    radar_3 = pyart.io.read_nexrad_archive(files_3[nframe])
    display_3 = pyart.graph.RadarDisplay(radar_3)
    display_3.plot_ppi('reflectivity', vmin = -32, vmax=80, cmap='pyart_NWSRef', title = files_3[nframe], colorbar_label='', axislabels_flag=False)

def animate_4(nframe):
    plt.clf()
    radar_4 = pyart.io.read_nexrad_archive(files_4[nframe])
    display_4 = pyart.graph.RadarDisplay(radar_4)
    display_4.plot_ppi('reflectivity', vmin = -32, vmax=80, cmap='pyart_NWSRef', title = files_4[nframe], colorbar_label='', axislabels_flag=False)

def animate_5(nframe):
    plt.clf()
    radar_5 = pyart.io.read_nexrad_archive(files_5[nframe])
    display_5 = pyart.graph.RadarDisplay(radar_5)
    display_5.plot_ppi('reflectivity', vmin = -32, vmax=80, cmap='pyart_NWSRef', title = files_5[nframe], colorbar_label='', axislabels_flag=False)

def animate_6(nframe):
    plt.clf()
    radar_6 = pyart.io.read_nexrad_archive(files_6[nframe])
    display_6 = pyart.graph.RadarDisplay(radar_6)
    display_6.plot_ppi('reflectivity', vmin = -32, vmax=80, cmap='pyart_NWSRef', title = files_6[nframe], colorbar_label='', axislabels_flag=False)


fig = plt.figure(figsize=(10,8))
radar_anim1 = animation.FuncAnimation(fig, animate_1, frames=len(files_1))
radar_anim1.save(filename = 'radar_animation_0-4hrs.gif', writer='imagemagick', fps=20)

fig = plt.figure(figsize=(10,8))
radar_anim2 = animation.FuncAnimation(fig, animate_2, frames=len(files_2))
radar_anim2.save(filename = 'radar_animation_4-8hrs.gif', writer='imagemagick', fps=20)

fig = plt.figure(figsize=(10,8))
radar_anim3 = animation.FuncAnimation(fig, animate_3, frames=len(files_3))
radar_anim3.save(filename = 'radar_animation_8-12hrs.gif', writer='imagemagick', fps=20)

fig = plt.figure(figsize=(10,8))
radar_anim4 = animation.FuncAnimation(fig, animate_4, frames=len(files_4))
radar_anim4.save(filename = 'radar_animation_12-16hrs.gif', writer='imagemagick', fps=20)

fig = plt.figure(figsize=(10,8))
radar_anim5 = animation.FuncAnimation(fig, animate_5, frames=len(files_5))
radar_anim5.save(filename = 'radar_animation_16-20hrs.gif', writer='imagemagick', fps=20)

fig = plt.figure(figsize=(10,8))
radar_anim6 = animation.FuncAnimation(fig, animate_6, frames=len(files_6))
radar_anim6.save(filename = 'radar_animation_20-24hrs.gif', writer='imagemagick', fps=20)
```
